```python
from google.colab.patches import cv2_imshow
from google.colab import drive
drive.mount('/content/drive')
import os
import cv2
import numpy as np
from tqdm.notebook import tqdm_notebook
```

```python
image_path = '/content/drive/MyDrive/PetColorization/images'
res = [300, 212]
data_x, data_y = [], []

for file_name in tqdm_notebook(os.listdir(image_path)[:]):
  try:
    img = cv2.imread(os.path.join(image_path, file_name))

    img = cv2.resize(img, (res[1], res[0]), interpolation=cv2.INTER_AREA)
    l_channel = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB))[:1]
    l_channel = np.array(l_channel)  # (1, 300, 212)

    img = cv2.resize(img, (res[1] // 16 * 4, res[0] // 16 * 4), interpolation=cv2.INTER_AREA)
    ab_channel = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB))[1:]
    ab_channel = np.array(ab_channel)  # (2, 72, 52)

    data_x.append(l_channel)
    data_y.append(ab_channel)
  except Exception as e:
    continue
```

```python
# Data is stored channel-first
# data[image #][channel][row][col]
print(np.array(data_x).shape)
print(np.array(data_y).shape)
print(f'X size: {np.array(data_x).nbytes / 1024 / 1024} MB')
print(f'Y size: {np.array(data_y).nbytes / 1024 / 1024} MB')
```

```python
np.save(f'/content/drive/MyDrive/PetColorization/data/data_x.npy', np.array(data_x))
np.save(f'/content/drive/MyDrive/PetColorization/data/data_y.npy', np.array(data_y))
```
